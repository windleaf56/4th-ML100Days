# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
```

```python
def ReLU(x):
    return abs(x) * (x>0)

def dReLU(x):
    return ( 1 * (x>0))
```

```python
ReLU(2)
```

```python
x = plt.linspace(-10,10,100)
```

```python
plt.plot(x, ReLU(x), 'b', label='linspace(-10,10,10)')
plt.grid()
plt.title('ReLU Function')
plt.text(-9, 9, r'$f(x)=Max(0,x)$', fontsize=15)
plt.show()
```

```python
plt.plot(x, dReLU(x), 'r', label='linspace(-10,10,10)')
plt.grid()
plt.title('dReLU Function')
plt.text(-9, 0.8, r'$f(x)=Max(0,x)$', fontsize=15)
plt.show()
```
